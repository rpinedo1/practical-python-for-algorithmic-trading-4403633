### Load the data

```python
import pandas as pd

df_microsoft = pd.read_excel(
    'data/Microsoft_Stock_Price_Historical_Daily.xlsx',
    parse_dates=['Date'], index_col=0
)

df_apple = pd.read_csv('data/Apple_Stock_Price_Historical_daily.csv',
    parse_dates=['Date'], index_col=0                     
)

# df_microsoft
df_apple
```

### Filter the data

![](src/Microsoft_x_LinkedIn.png)

```python
# Getting important rows of interest by using 
# loc[rows(''date':' to get rows starting at 'date),cols(':' to get all cols)]
df_apple_iphone = df_apple.loc['2007-01-01':,:].copy()
```

```python
df_apple_iphone
```

## Create new columns

### Tomorrow's percentage change

```python
# daily diff based on Close price, always use percentage
df_apple_iphone['change_tomorrow'] = df_apple_iphone.Close.pct_change(-1) * 100 * -1
```

```python
# dropping the last row since we don't have data for the next day
df_apple_iphone = df_apple_iphone.dropna().copy()
```

### Did the stock go up or down?

```python
import numpy as np
```

```python
df_apple_iphone['change_tomorrow_direction'] = np.where(df_apple_iphone.change_tomorrow > 0, 'UP', 'DOWN')
```

```python
df_apple_iphone.change_tomorrow_direction.value_counts()
```

## Export preprocessed DataFrame into an Excel

```python
df_apple_iphone.to_excel('data/Apple_Iphone_Processed.xlsx')
df_apple_iphone.to_csv('data/Apple_Iphone_Processed.csv')
```
